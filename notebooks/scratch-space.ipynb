```python
import numpy as np
import seaborn as sns

sns.set()
```

```python
def bounded_poisson(min_val, mean):
    # Shift distribution to change lower bound from 0 to min_val
    val = np.random.poisson(mean - min_val, size=1)[0] + min_val
    return val
```

```python
x = [bounded_poisson(1, 3) for _ in range(10000)]
sns.histplot(x, bins=20)
```

```python
x = np.random.poisson(lam=3, size=10000)
sns.histplot(x, bins=20)
```
